# DSP for Underwater Acoustic Positioning

Because RF does not progate well underwater, ROVs are not able to use GPS to find their location underwater.  One method to accomplish underwater position is through the use of acoustics.  Multiple transmitters on the surface send a pulse (or other signal), and the ROV can calulate its position based on the time of flight for these signals.  This document will look at the DSP needed for underwater acoustic positioning.
